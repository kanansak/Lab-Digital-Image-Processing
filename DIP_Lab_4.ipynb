```python
import cv2 as cv # import library opencv โดยย่อเป็น cv
import numpy as np # import library numpy โดยย่อเป็น np
from matplotlib import pyplot as plt  # import library matplotlib ในส่วนย่อย pyplot โดยย่อเป็น plt
from google.colab.patches import cv2_imshow  # import library google.colab  cv2_imshow
```

```python
!wget !wget https://assets.gezdev.com/dip/SUDOKU.png #ดาวน์โหลดไฟล์
```

```python
img = cv.imread("SUDOKU.png") #อ่านไฟล์ SUDOKU.png แล้วเก็บค่าในตัวแปร img
```

```python
cv2_imshow(img) #แสดงภาพ
```

```python
img_gray = cv.cvtColor(img,cv.COLOR_BGR2GRAY) #แปลงภาพเป็น BGR2GRAY
```

คำสั่ง

```python
_,imgNew = cv.threshold(img_gray,60,255,cv.THRESH_BINARY) #ทำการปรับค่า threshold (ภาพสีเทา,ค่าที่ต้องการปรับ,ค่าสุงสุด,ประเภทของ threshold)
```

```python
cv2_imshow(imgNew) #แสดงภาพ
```

```python
_,img_bininv = cv.threshold(img_gray,70,255,cv.THRESH_BINARY_INV) #ทำการปรับค่า threshold (ภาพสีเทา,ค่าที่ต้องการปรับ,ค่าสุงสุด,ประเภทของ threshold) 
#THRESH_BINARY_INV ค่าตรงข้าม
```

```python
cv2_imshow(img_bininv) #แสดงภาพ
```

```python

plt.hist(img_gray.ravel(),256,[0,256]); plt.show() #แสดง Histrogram จากตัวแปร img_gray
```

# ทดลองปรับเปลี่ยนค่า

```python
#ใส่ค่า threshold น้อยไป
_,imgNew = cv.threshold(img_gray,30,255,cv.THRESH_BINARY) 
#ภาพที่ได้มีการขาดหายไป

#ใส่ค่า threshold มากไป
_,imgNew2 = cv.threshold(img_gray,100,255,cv.THRESH_BINARY) 
#ภาพที่ได้มีสีดำ ทำให้ภาพบางส่วนขาดหายไป
res = np.hstack((img_gray,imgNew,imgNew2)) #ภาพเรียงต่อกันแนวนอน
cv2_imshow(res) #แสดงภาพ
```

```python
#ใส่ค่า threshold น้อยไป
_,img_bininv = cv.threshold(img_gray,40,255,cv.THRESH_BINARY_INV)
#ภาพที่ได้มีการขาดหายไป

#ใส่ค่า threshold มากไป
_,img_bininv2 = cv.threshold(img_gray,120,255,cv.THRESH_BINARY_INV)
#ภาพที่ได้มีสีดำ ทำให้ภาพบางส่วนขาดหายไป
res = np.hstack((img_gray,img_bininv,img_bininv2)) #ภาพเรียงต่อกันแนวนอน
cv2_imshow(res) #แสดงภาพ
```
