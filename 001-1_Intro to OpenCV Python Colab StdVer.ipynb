##how to display the images
* In this step we will read images from urls, and display them using openCV, 
* please note the difference when reading image in RGB and BGR format. 
* The default input color channels are in BGR format for openCV.

####Method 1 using cv2_imshow

```python
# Method 1 using cv2_imshow -> from google.colab.patches import cv2_imshow # for image display
cv2_imshow(cat_img)
cv2_imshow(my_img)
```

###Method 2 using plt.imshow matplotlib

```python
# 2 using plt.imshow matplotlib
# convert bgr to rgb
rgr_cat_img = cv2.cvtColor(cat_img,cv2.COLOR_BGR2RGB)
rgr_my_img = cv2.cvtColor(my_img,cv2.COLOR_BGR2RGB)
# concatnate images
compare_cat_imag=cv2.hconcat([cat_img,rgr_cat_img])
compare_my_image=cv2.hconcat([my_img,rgr_my_img])
plt.imshow(compare_cat_imag)
plt.show()
plt.imshow(compare_my_image)
plt.show()
```

### Resize image 

```python
#shape[0]=>ความกว้าง
#shape[1]=>ความสูง
#shape[2]=> จำนวน channel 
#print('shape before')
#print(cat_img.shape)
#print(my_img.shape)
cat_imgRS = cv2.resize(cat_img,(int(cat_img.shape[1]*.25),int(cat_img.shape[0]*.25)))
my_imgRS = cv2.resize(my_img,(int(my_img.shape[1]/4),int(my_img.shape[0]/4)))
#cv2_imshow(cat_img)
#cv2_imshow(my_img)
#print('shape after')
#print(cat_img.shape)
#print(my_img.shape)
```

```python
print('shape before')
print(cat_img.shape)
print(my_img.shape)
cv2_imshow(cat_img)
cv2_imshow(cat_imgRS)
cv2_imshow(my_img)
cv2_imshow(my_imgRS)
print('shape after')
print(cat_img.shape)
print(my_img.shape)
```

### Flip

```python
filped_cat_img = cv2.flip(cat_img,1)
cv2_imshow(cat_img)
cv2_imshow(filped_cat_img)
```

# putText 

```python
cv2.putText(filped_cat_img,"Cat",(200,150),cv2.FONT_HERSHEY_SIMPLEX, 1.5,(255,0,0),2)
cv2_imshow(filped_cat_img)
```

### imwrite

```python
ret_write = cv2.imwrite("cat.jpg",filped_cat_img) # PNG / BMP
ret_write
```

### Color 2 GrayScale

```python
gray_filped_cat_img = cv2.cvtColor(filped_cat_img,cv2.COLOR_BGR2GRAY)
cv2_imshow(gray_filped_cat_img)
```
