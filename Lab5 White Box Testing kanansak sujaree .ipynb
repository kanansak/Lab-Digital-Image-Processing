```python
def ElectricityBillCalculator(unit,over5A):
  result = 0 #ค่าไฟที่ต้องจ่าย
  if (over5A==False):
    if(unit < 150):
      if(unit < 16):
        result = unit*2.3488
      elif(unit < 26):
        result = unit*2.9882
      elif(unit < 36):
        result = unit*3.2405
      elif(unit < 101):
        result = unit*3.6237
      else:
        result = unit*3.7171
      result += 8.19
    else:
      if(unit < 401):
        result = unit*4.2218
      else:
        result = unit*4.4217
      result += 38.22
  else: #over5A==True
     if(unit < 401):
        result = unit*4.2218
     else:
        result = unit*4.4217

     result += 38.22

  return result
```
